```python
csv_file = "WB_Data.csv"
travel_df = pd.read_csv(csv_file)
travel_df.head()
```

```python
travel = travel_df.drop(columns=['Series Name', 'Series Code'])
travel.head()
```

```python
tra_df = travel.melt(id_vars=["Country Name", "Country Code"], var_name="Departure Year")
tra_df = tra_df.rename(columns={"value": "No. Of Travellers"})
tra_df
```

```python
tra_df["Departure Year"] = tra_df["Departure Year"].str.slice(0,4)
tra_df 
```

```python
clean_df = tra_df.sort_values(by=['Country Name','Departure Year'])
clean_df
```
